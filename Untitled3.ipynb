## Data cleaning

### Subtask:
Clean the twitter archive dataset.


**Reasoning**:

````python
# Handling missing values
# For now, fill missing values in numerical columns with the mean
numerical_cols = ['rating_numerator', 'rating_denominator']
for col in numerical_cols:
    df_twitter_archive[col] = df_twitter_archive[col].fillna(df_twitter_archive[col].mean())

# Fill missing values in categorical columns with 'Unknown'
categorical_cols = ['name', 'doggo', 'floofer', 'pupper', 'puppo']
for col in categorical_cols:
    df_twitter_archive[col] = df_twitter_archive[col].fillna('Unknown')

# Remove duplicate rows, keeping the first occurrence
df_twitter_archive.drop_duplicates(inplace=True)

# Address inconsistencies
# Check and handle invalid values (e.g., negative ratings)
df_twitter_archive = df_twitter_archive[df_twitter_archive['rating_numerator'] >= 0]
df_twitter_archive = df_twitter_archive[df_twitter_archive['rating_denominator'] > 0]

# Display cleaned DataFrame
display(df_twitter_archive.head())
print(df_twitter_archive.shape)
````

````python
# Import necessary libraries
import pandas as pd
import json
import tweepy
from tweepy import OAuthHandler
from timeit import default_timer as timer
import sqlite3
from sqlalchemy import create_engine # Used for optional database storage
from IPython.display import display # For displaying data in Jupyter

# --- Configuration ---
# Replace 'HIDDEN' with your actual Twitter API credentials
# Ensure tweepy is installed: !pip install tweepy==3.10.0
CONSUMER_KEY = 'HIDDEN'
CONSUMER_SECRET = 'HIDDEN'
ACCESS_TOKEN = 'HIDDEN'
ACCESS_SECRET = 'HIDDEN'

# File paths
TWITTER_ARCHIVE_FILE = 'twitter-archive-enhanced.csv'
IMAGE_PREDICTIONS_FILE = 'image-predictions.tsv'
TWEET_JSON_FILE = 'tweet_json.txt' # This file will be created/used

# --- 1. Data Loading ---
print("--- Starting Data Loading ---")

# Load twitter-archive-enhanced.csv
print(f"\nLoading {TWITTER_ARCHIVE_FILE}...")
try:
    df_twitter_archive = pd.read_csv(TWITTER_ARCHIVE_FILE, encoding='utf-8')
    print(f"Loaded {TWITTER_ARCHIVE_FILE} with utf-8 encoding.")
except UnicodeDecodeError:
    try:
        df_twitter_archive = pd.read_csv(TWITTER_ARCHIVE_FILE, encoding='latin-1')
        print(f"Loaded {TWITTER_ARCHIVE_FILE} with latin-1 encoding.")
    except FileNotFoundError:
        print(f"Error: {TWITTER_ARCHIVE_FILE} not found.")
        df_twitter_archive = None
    except Exception as e:
        print(f"Error loading {TWITTER_ARCHIVE_FILE}: {e}")
        df_twitter_archive = None

if df_twitter_archive is not None:
    print(f"{TWITTER_ARCHIVE_FILE} loaded successfully. Shape: {df_twitter_archive.shape}")
    display(df_twitter_archive.head())
else:
    print(f"Failed to load {TWITTER_ARCHIVE_FILE}. Subsequent steps may be affected.")


# Load image-predictions.tsv
print(f"\nLoading {IMAGE_PREDICTIONS_FILE}...")
try:
    # Using '\s+' as separator for multiple spaces and engine='python'
    df_image_predictions = pd.read_csv(IMAGE_PREDICTIONS_FILE, sep='\s+', engine='python')
    print(f"Loaded {IMAGE_PREDICTIONS_FILE} successfully.")
except FileNotFoundError:
    print(f"Error: {IMAGE_PREDICTIONS_FILE} not found.")
    df_image_predictions = None
except Exception as e:
    print(f"Error loading {IMAGE_PREDICTIONS_FILE}: {e}")
    df_image_predictions = None

if df_image_predictions is not None:
    print(f"{IMAGE_PREDICTIONS_FILE} loaded successfully. Shape: {df_image_predictions.shape}")
    display(df_image_predictions.head())
else:
    print(f"Failed to load {IMAGE_PREDICTIONS_FILE}. Subsequent steps may be affected.")

# Load tweet_json.txt (from Twitter API)
# First, check if the file exists. If not, attempt to query the API.
print(f"\nLoading {TWEET_JSON_FILE}...")
df_tweet_json = None # Initialize to None

try:
    with open(TWEET_JSON_FILE, 'r') as f:
        tweet_data = []
        for line in f:
            if line.strip(): # Skip empty lines
                tweet_data.append(json.loads(line))
    df_tweet_json = pd.DataFrame(tweet_data)
    print(f"Loaded {TWEET_JSON_FILE} successfully.")

except FileNotFoundError:
    print(f"{TWEET_JSON_FILE} not found. Attempting to query Twitter API...")
    if df_twitter_archive is not None and 'tweet_id' in df_twitter_archive.columns:
        # Ensure tweepy is installed before querying
        try:
            auth = OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
            auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
            api = tweepy.API(auth, wait_on_rate_limit=True)

            tweet_ids = df_twitter_archive.tweet_id.values.tolist() # Convert to list
            print(f"Number of tweet IDs to query: {len(tweet_ids)}")

            count = 0
            fails_dict = {}
            start = timer()

            try:
                with open(TWEET_JSON_FILE, 'w') as outfile:
                    for tweet_id in tweet_ids:
                        count += 1
                        # print(f"{count}: Querying tweet ID {tweet_id}") # Optional: print progress
                        try:
                            # Use get_status with tweet_mode='extended' for full text
                            tweet = api.get_status(tweet_id, tweet_mode='extended')
                            json.dump(tweet._json, outfile)
                            outfile.write('\n')
                        except tweepy.TweepError as e:
                            fails_dict[tweet_id] = e
                            # print(f"API Error for tweet {tweet_id}: {e}") # Optional: print errors
                            pass # Continue to the next tweet ID
                end = timer()
                print(f"Twitter API query finished in {end - start:.2f} seconds.")
                print(f"Failed to retrieve data for {len(fails_dict)} tweets.")
                # Optionally print fails_dict for details: print(fails_dict)

                # Now attempt to load the newly created file
                try:
                     with open(TWEET_JSON_FILE, 'r') as f:
                        tweet_data = []
                        for line in f:
                            if line.strip():
                                tweet_data.append(json.loads(line))
                     df_tweet_json = pd.DataFrame(tweet_data)
                     print(f"Loaded {TWEET_JSON_FILE} successfully after API query.")
                except Exception as e:
                    print(f"Error loading {TWEET_JSON_FILE} after API query: {e}")
                    df_tweet_json = None

            except Exception as e:
                print(f"An error occurred during Twitter API query or writing to {TWEET_JSON_FILE}: {e}")
                df_tweet_json = None # Ensure df_tweet_json is None on failure

        except ImportError:
            print("Tweepy is not installed. Skipping Twitter API query. Please install tweepy (`!pip install tweepy==3.10.0`).")
            df_tweet_json = None
        except Exception as e:
             print(f"An error occurred during Twitter API setup or initial query attempt: {e}")
             df_tweet_json = None
    else:
        print(f"Cannot query Twitter API because {TWITTER_ARCHIVE_FILE} was not loaded or is missing 'tweet_id'.")
        df_tweet_json = None # Ensure df_tweet_json is None if prerequisite is missing

except json.JSONDecodeError as e:
    print(f"Error decoding JSON from {TWEET_JSON_FILE}: {e}")
    df_tweet_json = None
except Exception as e:
    print(f"An unexpected error occurred while loading {TWEET_JSON_FILE}: {e}")
    df_tweet_json = None


if df_tweet_json is not None:
    print(f"{TWEET_JSON_FILE} loaded successfully. Shape: {df_tweet_json.shape}")
    display(df_tweet_json.head())
else:
    print(f"Failed to load or generate {TWEET_JSON_FILE}. Subsequent steps may be affected.")

print("--- Data Loading Complete ---")


# --- 2. Data Cleaning and Preparation ---
print("\n--- Starting Data Cleaning and Preparation ---")

# Clean df_twitter_archive
if df_twitter_archive is not None:
    print("\nCleaning df_twitter_archive...")
    # Fill missing numerical columns with the mean
    numerical_cols_archive = ['rating_numerator', 'rating_denominator']
    for col in numerical_cols_archive:
        if col in df_twitter_archive.columns and df_twitter_archive[col].isnull().any():
             df_twitter_archive[col] = df_twitter_archive[col].fillna(df_twitter_archive[col].mean())

    # Fill missing categorical columns with 'Unknown'
    categorical_cols_archive = ['name', 'doggo', 'floofer', 'pupper', 'puppo']
    for col in categorical_cols_archive:
        if col in df_twitter_archive.columns and df_twitter_archive[col].isnull().any():
            df_twitter_archive[col] = df_twitter_archive[col].fillna('Unknown')

    # Remove duplicate rows
    initial_rows = df_twitter_archive.shape[0]
    df_twitter_archive.drop_duplicates(inplace=True)
    if df_twitter_archive.shape[0] < initial_rows:
        print(f"Removed {initial_rows - df_twitter_archive.shape[0]} duplicate rows from df_twitter_archive.")

    # Address inconsistencies (negative ratings)
    if 'rating_numerator' in df_twitter_archive.columns and 'rating_denominator' in df_twitter_archive.columns:
        initial_rows = df_twitter_archive.shape[0]
        df_twitter_archive = df_twitter_archive[df_twitter_archive['rating_numerator'] >= 0]
        df_twitter_archive = df_twitter_archive[df_twitter_archive['rating_denominator'] > 0]
        if df_twitter_archive.shape[0] < initial_rows:
             print(f"Removed {initial_rows - df_twitter_archive.shape[0]} rows with invalid ratings from df_twitter_archive.")

    # Convert tweet_id to appropriate type for merging (e.g., int64)
    if 'tweet_id' in df_twitter_archive.columns:
        try:
            df_twitter_archive['tweet_id'] = df_twitter_archive['tweet_id'].astype('int64')
            print("Converted 'tweet_id' in df_twitter_archive to int64.")
        except Exception as e:
            print(f"Could not convert 'tweet_id' in df_twitter_archive to int64: {e}")

    print("df_twitter_archive cleaned. Shape:", df_twitter_archive.shape)
    display(df_twitter_archive.head())
else:
    print("df_twitter_archive is None. Skipping cleaning.")


# Clean df_image_predictions
if df_image_predictions is not None:
    print("\nCleaning df_image_predictions...")
    # Convert confidence columns to numeric, coercing errors to NaN
    conf_cols = ['p1_conf', 'p2_conf', 'p3_conf']
    for col in conf_cols:
        if col in df_image_predictions.columns:
            df_image_predictions[col] = pd.to_numeric(df_image_predictions[col], errors='coerce')
            # Optional: Fill NaN confidence scores if appropriate, e.g., with 0
            # df_image_predictions[col] = df_image_predictions[col].fillna(0)


    # Remove duplicate rows
    initial_rows = df_image_predictions.shape[0]
    # Use .copy() after filtering to avoid SettingWithCopyWarning
    df_image_predictions = df_image_predictions[~df_image_predictions.duplicated()].copy()
    if df_image_predictions.shape[0] < initial_rows:
        print(f"Removed {initial_rows - df_image_predictions.shape[0]} duplicate rows from df_image_predictions.")

    # Convert tweet_id to appropriate type for merging (e.g., int64)
    if 'tweet_id' in df_image_predictions.columns:
        try:
            df_image_predictions['tweet_id'] = df_image_predictions['tweet_id'].astype('int64')
            print("Converted 'tweet_id' in df_image_predictions to int64.")
        except Exception as e:
            print(f"Could not convert 'tweet_id' in df_image_predictions to int64: {e}")


    print("df_image_predictions cleaned. Shape:", df_image_predictions.shape)
    display(df_image_predictions.head())
else:
     print("df_image_predictions is None. Skipping cleaning.")


# Clean df_tweet_json
if df_tweet_json is not None:
    print("\nCleaning df_tweet_json...")
    # Remove duplicate rows
    initial_rows = df_tweet_json.shape[0]
    df_tweet_json.drop_duplicates(inplace=True)
    if df_tweet_json.shape[0] < initial_rows:
         print(f"Removed {initial_rows - df_tweet_json.shape[0]} duplicate rows from df_tweet_json.")

    # Keep only relevant columns before merging
    columns_to_keep_json = ['id', 'created_at', 'favorite_count', 'retweet_count', 'full_text', 'source']
    # Check which columns are actually in df_tweet_json
    valid_columns_to_keep_json = [col for col in columns_to_keep_json if col in df_tweet_json.columns]
    # Use .copy() after selecting columns to avoid SettingWithCopyWarning
    df_tweet_json_cleaned = df_tweet_json[valid_columns_to_keep_json].copy()
    print(f"Keeping columns in df_tweet_json: {valid_columns_to_keep_json}")

    # Rename 'id' to 'tweet_id' for merging
    if 'id' in df_tweet_json_cleaned.columns:
        df_tweet_json_cleaned.rename(columns={'id': 'tweet_id'}, inplace=True)
        print("Renamed 'id' to 'tweet_id' in df_tweet_json.")

    # Convert tweet_id to appropriate type for merging (e.g., int64)
    if 'tweet_id' in df_tweet_json_cleaned.columns:
        try:
            df_tweet_json_cleaned['tweet_id'] = df_tweet_json_cleaned['tweet_id'].astype('int64')
            print("Converted 'tweet_id' in df_tweet_json to int64.")
        except Exception as e:
            print(f"Could not convert 'tweet_id' in df_tweet_json to int64: {e}")

    # Convert created_at to datetime
    if 'created_at' in df_tweet_json_cleaned.columns:
        # Ensure it's string before converting
        df_tweet_json_cleaned['created_at'] = df_tweet_json_cleaned['created_at'].astype(str)
        # Use errors='coerce' to turn unparseable dates into NaT (Not a Time)
        df_tweet_json_cleaned['created_at'] = pd.to_datetime(df_tweet_json_cleaned['created_at'], errors='coerce')
        print("Converted 'created_at' in df_tweet_json to datetime.")

    print("df_tweet_json cleaned. Shape:", df_tweet_json_cleaned.shape)
    display(df_tweet_json_cleaned.head())
else:
    print("df_tweet_json is None. Skipping cleaning.")
    df_tweet_json_cleaned = None # Ensure this is None if source was None


print("--- Data Cleaning and Preparation Complete ---")


# --- 3. Merge DataFrames ---
print("\n--- Starting Data Merging ---")
df_master = None # Initialize master DataFrame

# Merge df_twitter_archive and df_image_predictions first
if df_twitter_archive is not None and df_image_predictions is not None:
    if 'tweet_id' in df_twitter_archive.columns and 'tweet_id' in df_image_predictions.columns:
        try:
             # Check data types before merging and inform user if they are different
             if df_twitter_archive['tweet_id'].dtype != df_image_predictions['tweet_id'].dtype:
                 print(f"Warning: 'tweet_id' data types differ between archive ({df_twitter_archive['tweet_id'].dtype}) and image predictions ({df_image_predictions['tweet_id'].dtype}). Attempting merge.")
                 # If needed, explicitly convert one to match the other, e.g.:
                 # df_image_predictions['tweet_id'] = df_image_predictions['tweet_id'].astype(df_twitter_archive['tweet_id'].dtype)


             # Use 'left' merge to keep all tweets from the archive file as the base
             df_merged = pd.merge(df_twitter_archive, df_image_predictions, on='tweet_id', how='left', suffixes=('_archive', '_image'))
             print("Merged df_twitter_archive and df_image_predictions (left merge).")
             print("Shape after first merge:", df_merged.shape)
        except Exception as e:
            print(f"Error during first merge (archive and images): {e}")
            df_merged = None # Ensure df_merged is None on merge failure
    else:
        print("'tweet_id' column missing in one or both of df_twitter_archive or df_image_predictions. Cannot perform first merge.")
        df_merged = None # Ensure df_merged is None if columns are missing
else:
    print("df_twitter_archive or df_image_predictions is None. Skipping first merge.")
    df_merged = None # Ensure df_merged is None if source was None


# Merge the result with df_tweet_json_cleaned
if df_merged is not None and df_tweet_json_cleaned is not None:
    if 'tweet_id' in df_merged.columns and 'tweet_id' in df_tweet_json_cleaned.columns:
         try:
             # Check data types before merging and inform user if they are different
             if df_merged['tweet_id'].dtype != df_tweet_json_cleaned['tweet_id'].dtype:
                  print(f"Warning: 'tweet_id' data types differ between first merge result ({df_merged['tweet_id'].dtype}) and tweet json ({df_tweet_json_cleaned['tweet_id'].dtype}). Attempting merge.")
                  # If needed, explicitly convert one to match the other
                  # df_tweet_json_cleaned['tweet_id'] = df_tweet_json_cleaned['tweet_id'].astype(df_merged['tweet_id'].dtype)

             # Use 'left' merge to keep all tweets from the previous merge result
             # Ensure we handle potential column name conflicts, though with cleaned_json subset this should be minimal.
             # Suffixes are added just in case, although 'full_text' and 'source' will be the same source.
             df_master = pd.merge(df_merged, df_tweet_json_cleaned, on='tweet_id', how='left', suffixes=('_prev', '_json'))

             # Optional: Resolve columns with duplicate names if necessary (e.g., '_prev' and '_json')
             # Based on the cleaning, 'full_text' and 'source' from json should overwrite/replace
             # the potentially less complete versions from the original archive if needed.
             # Since we are doing a left merge and keeping columns from json, the json columns should take precedence.
             # We can drop the old ones if they exist:
             columns_to_drop_after_merge = ['full_text_prev', 'source_prev', 'created_at_prev'] # Add other columns if they got suffixes
             for col in columns_to_drop_after_merge:
                 if col in df_master.columns:
                     df_master.drop(columns=[col], inplace=True)
                     # Rename the json versions if they got suffixes
                     if f"{col}_json" in df_master.columns: # Example check
                         df_master.rename(columns={f"{col}_json": col}, inplace=True) # Example rename

             print("Merged with df_tweet_json_cleaned (left merge).")
             print("Shape of the final master DataFrame:", df_master.shape)
             print("Columns in final master DataFrame:", df_master.columns.tolist())

         except Exception as e:
             print(f"Error during second merge (with tweet json): {e}")
             df_master = None # Ensure df_master is None on merge failure
    else:
         print("'tweet_id' column missing in one or both of the first merge result or df_tweet_json_cleaned. Cannot perform second merge.")
         df_master = None # Ensure df_master is None if columns are missing
else:
    print("df_merged or df_tweet_json_cleaned is None. Skipping second merge.")
    df_master = None # Ensure df_master is None if source was None


if df_master is not None:
    print("\nFinal Master DataFrame head:")
    display(df_master.head())
    print("--- Data Merging Complete ---")
else:
    print("\n--- Data Merging Failed. Master DataFrame is None. ---")


# --- 4. Storing Data ---
print("\n--- Starting Data Storage ---")

# Store the cleaned master DataFrame in a CSV file
CSV_OUTPUT_FILE = 'twitter_archive_master.csv'
if df_master is not None:
    try:
        df_master.to_csv(CSV_OUTPUT_FILE, index=False)
        print(f"Cleaned master DataFrame saved to {CSV_OUTPUT_FILE}")
    except Exception as e:
        print(f"Error saving master DataFrame to CSV: {e}")
else:
    print("Master DataFrame is None. Cannot save to CSV.")


# Optionally store the cleaned data in a SQLite database
DATABASE_NAME = 'twitter_archive_master.db'
if df_master is not None:
    try:
        # Create a SQLite engine
        engine = create_engine(f'sqlite:///{DATABASE_NAME}')

        # Connect to the database and store the DataFrame
        df_master.to_sql('twitter_archive', engine, index=False, if_exists='replace')

        print(f"\nCleaned master DataFrame stored in SQLite database: {DATABASE_NAME}")

        # Verify by reading from the database
        with engine.connect() as conn:
            try:
                 df_from_db = pd.read_sql("SELECT * FROM twitter_archive LIMIT 5", conn)
                 print("\nData read from database (first 5 rows):")
                 display(df_from_db)
            except Exception as e:
                 print(f"Error reading from SQLite database: {e}")


    except ImportError:
        print("\nSQLAlchemy is not installed. Skipping SQLite database storage.")
    except Exception as e:
        print(f"\nError storing master DataFrame in SQLite database: {e}")
else:
    print("Master DataFrame is None. Skipping SQLite database storage.")

print("--- Data Storage Complete ---")

# --- End of Script ---
````
